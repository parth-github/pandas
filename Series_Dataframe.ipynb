# Pandas- Series

```python
import pandas as pd
import numpy as np
```

```python
# List
col_idx = ['a', 'b', 'c'] # Index Column
col_val = [20, 30, 40]

# Numpy Array
arr = np.array(col_val)

# Dictionary
d = {
    'idx_row1': 10,
    'idx_row2': 30,
    'idx_row3': 40
}

```

```python
# @title Create Series from list
pd.Series(col_val)
pd.Series(data=col_val, index=col_idx, dtype=float)
#pd.Series(col_val, col_idx )
```

```python
# @title Create Series from Numpy Array

pd.Series(arr)
pd.Series(arr, col_idx)
```

```python
# @title Create Series from Dictionary
# Key is index
# Value is data

pd.Series(d)
```

# Multi index

```python
outside_idx = ['G1', 'G1', 'G1', 'G2', 'G2', 'G2']
inside_idx = [1, 2, 3, 1, 2, 3]
multi_idx = list(zip(outside_idx, inside_idx))

multi_idx = pd.MultiIndex.from_tuples(multi_idx)
```

```python
df_multi_idx = pd.DataFrame(randn(6,5), multi_idx, ['col1','col2', 'col3', 'col4', 'col5'])
df_multi_idx
```

```python
# @title Call multi-leve index dataframe
df_multi_idx.loc['G1']
df_multi_idx.loc['G1']
```

```python
df_multi_idx.index.names
```

```python
df_multi_idx.index.names = ['Group', 'Team']
df_multi_idx.index.names
```

```python
df_multi_idx
```

```python
df_multi_idx.xs('G1')
```

```python
df_multi_idx.xs(1, level='Team')
```
