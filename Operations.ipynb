```python
# @title Pivot

data = {
    'A': ['foo', 'foo', 'foo', 'bar', 'bar', 'bar'],
    'B': ['one', 'one', 'two', 'two', 'one', 'one'],
    'C': ['x', 'y', 'x', 'y', 'x', 'y'],
    'D': [1, 2, 3, 4, 5, 6]
}

df2 = pd.DataFrame(data)
df2
```

```python
df2. \
pivot_table(
    values='D',
    index=['A', 'B'],
    columns=['C'],
    aggfunc='count')
```

```python
df2. \
pivot_table(
    values='D',
    index=['A', 'B'],
    columns=['C'],
    aggfunc='sum')
```
